```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sfm
import statsmodels.api as sm
from sklearn.metrics import roc_curve, auc
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 500)
```

```python
# Lectura de archivo de Excel

filename = 'C:/Users/sandra.pastran/Desktop/1.Diplomado/Telefonia/Clientes19_20.xlsx'

df_ClientesAgos19 = pd.read_excel(filename, 
                                  sheet_name='Agosto2019')

df_ClientesAgos19.head()
```

```python
df_ClientesAgos19.info()
```

```python
df_ClientesAgos19['aux_pie'] = 1
df_ClientesAgos19.aux_pie.groupby(df_ClientesAgos19.Plan).sum().plot(kind='pie', autopct='%1.0f%%')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot('N_llamadas', 'Antiguedad', size='Ciudad', data=df_ClientesAgos19)
#sns.scatterplot('N_llamadas', 'Valor', hue='Plan',style='grupo',size='Ciudad', data=df_ClientesAgos19)
plt.show()
```

```python
corr_matrix = df_ClientesAgos19[['Tipo','Antiguedad','Plan','Edad','Sexo','Procedencia','Estudios','grupo','Status','Ingresos','Ciudad','N_llamadas','Valor']].corr()
```

```python
sns.heatmap(corr_matrix,cmap=sns.diverging_palette(220, 10, n=50),annot=True)
plt.title("Correlation Matrix")
```

```python
df_ClientesAgos19['Intercept'] = 1
logit = sm.Logit(df_ClientesAgos19['CHURN'], df_ClientesAgos19[['Intercept','Ciudad','Edad','Sexo','Procedencia','Estudios','Ingresos','N_llamadas','Antiguedad']])
#logit = sm.Logit(df_ClientesAgos19['CHURN'], df_ClientesAgos19[['Intercept','Tipo','Antiguedad','Plan','Edad','Sexo','Procedencia', 'N_llamadas']])
logit_res = logit.fit()
logit_res.summary()
```
